```python
# We don't need the most-inner loop. 
# As long as the total up to that point is less or equal to 1, 
# we can count the pennies (1 cent) using subtraction.
combination_amounts = []
total_so_far  = 0

for amount_100 in range(0, 101, 100):
  for amount_50 in range(0, 101, 50):
    for amount_25 in range(0, 101, 25):
      for amount_10 in range(0, 101, 10):
        for amount_5 in range(0, 101, 5):
          total_so_far = amount_100 + amount_50 + amount_25 + amount_10 + amount_5

          if total_so_far <= 100:
            combination_amounts.append([amount_100, amount_50, amount_25, amount_10, amount_5, 100 - total_so_far])
```

```python
# [copy code from above]

def find_combinations(total):
    combinations_amounts = []

    for amount_100 in range(0, total+1, 100):
        for amount_50 in range(0, total+1, 50):
            for amount_25 in range(0, total+1, 25):
                for amount_10 in range(0, total+1, 10):
                    for amount_5 in range(0, total+1, 5):
                        total_so_far = amount_100 + amount_50 + amount_25 + amount_10 + amount_5

                        if total_so_far <= total:
                            combinations_amounts.append([amount_100, amount_50, amount_25, amount_10, amount_5,
                                                         total - total_so_far])
    
    return combinations_amounts
```

```python
def find_combinations_chau(given_amount):
  end = given_amount + 1
  combination_amounts = []
  total_so_far = 0

  for amount_100 in range(0, end, 100):
    for amount_50 in range(0, end, 50):
      for amount_25 in range(0, end, 25):
        for amount_10 in range(0, end, 10):
          for amount_5 in range(0, end, 5):
            total_so_far = amount_100 + amount_50 + amount_25 + amount_10 + amount_5

            if(total_so_far <= given_amount):
              combination_amounts.append([amount_100, amount_50, amount_25, amount_10, amount_5, given_amount - total_so_far])
  
  return combination_amounts
```
